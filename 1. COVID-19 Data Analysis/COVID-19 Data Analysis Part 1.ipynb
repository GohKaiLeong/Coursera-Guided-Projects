# COVID-19 Data Analysis Part 1

Import the libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Import the dataset of confirmed COVID-19 cases

File is available at https://data.humdata.org/dataset/novel-coronavirus-2019-ncov-cases

```python
corona_dataset_csv = pd.read_csv("covid19_confirmed_global.csv")
print(corona_dataset_csv.shape)
corona_dataset_csv.head(10)
```

### Delete useless columns

```python
df1 = corona_dataset_csv.drop(["Lat", "Long"], axis = 1)
df1.head(10)
```

### Aggregating the rows by country

```python
df1_aggregated = df1.groupby("Country/Region").sum()
print(df1_aggregated.shape)
df1_aggregated.head(10)
```

### Visualize data related to a particular country (e.g. China)

```python
df1_aggregated.loc["China"]
```

### Calculate a good measurement
We need a good measurement in numerical form, describing the spread of virus in a country

#### Countries affected earlier

```python
df1_aggregated.loc["China"].plot()
df1_aggregated.loc["Italy"].plot()
df1_aggregated.loc["Spain"].plot()
plt.legend()
```

#### Countries affected later

```python
df1_aggregated.loc["US"].plot()
df1_aggregated.loc["Brazil"].plot()
df1_aggregated.loc["India"].plot()
plt.legend()
```

### Plot first derivative curve

#### Countries affected earlier

```python
df1_aggregated.loc["China"].diff().plot()
df1_aggregated.loc["Italy"].diff().plot()
df1_aggregated.loc["Spain"].diff().plot()
plt.legend()
```

#### Countries affected later

```python
df1_aggregated.loc["US"].diff().plot()
df1_aggregated.loc["Brazil"].diff().plot()
df1_aggregated.loc["India"].diff().plot()
plt.legend()
```

### Find max infection rates

```python
print("China's max infection:", df1_aggregated.loc["China"].diff().max())
print("Italy's max infection:", df1_aggregated.loc["Italy"].diff().max())
print("Spain's max infection:", df1_aggregated.loc["Spain"].diff().max())
print("US's max infection:", df1_aggregated.loc["US"].diff().max())
print("Brazil's max infection:", df1_aggregated.loc["Brazil"].diff().max())
print("India's max infection:", df1_aggregated.loc["India"].diff().max())
```

#### For all the countries

```python
countries = list(df1_aggregated.index)
max_infect = []
for c in countries:
    max_infect.append(df1_aggregated.loc[c].diff().max())
df1_aggregated["max_infection"] = max_infect

df1_aggregated.head()
```

### New dataframe with only needed columns

```python
corona_data = pd.DataFrame(df1_aggregated["max_infection"])
corona_data.head()
```

### Import dataset for World Happiness report

Available at https://www.kaggle.com/unsdsn/world-happiness/data

```python
happy = pd.read_csv("world_happiness_2019.csv")
happy.head()
```

### Drop the useless columns

```python
useless = ["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
happy.drop(useless, axis=1, inplace=True)
happy.head()
```

### Changing the index of the dataframe

```python
happy.set_index("Country or region", inplace=True)
happy.head()
```

### Join the dataset

```python
data = corona_data.join(happy, how="inner")
data.head()
```

### Correlation matrix

```python
data.corr()
```
